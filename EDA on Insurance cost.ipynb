# Data visualization

# Univariate Analysis 

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.distplot(df.charges,color='r')
plt.title('Charge Distribution', size=18)
plt.xlabel('Charges',size=14)
plt.ylabel('Density',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df.age,color='y')
plt.title('Age Distribution', size=18)
plt.xlabel('Age',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='sex',data=df)
plt.title('Total Number of Male and Female', size=18)
plt.xlabel('Sex',size=14)
plt.show()
```

# Bivariate Analysis

```python
# Age vs Charges
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age',y='charges',color='g',data=df)
plt.title('Age vs Charges',size=18)
plt.xlabel('Age',size=14)
plt.ylabel('Charges',size=14)
plt.show()
# Scatterplot shows the relationship between two numerical variables
```

```python
#Smoker vs Charges
plt.figure(figsize=(10,6))
sns.boxplot(x='smoker',y='charges',data=df)
plt.title('Smoker vs Charges',size=18)
#This represent the cost of insurance for smokers is higher than for non=smoker
```

```python
# Pairplot shows the relationship between 2 numerical values as well as distribution of each variable
sns.pairplot(df,
            markers="+",
            diag_kind="kde",
            kind='reg',
            plot_kws={'line_kws':{'color':'#aec6cf'},
                     'scatter_kws':{'alpha':0.7,
                                   'color':'red'}},
            corner=True);
```

# Multivariate Analysis

```python
numerical_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,6))
sns.heatmap(numerical_df.corr(),
annot=True,square=True)
plt.title('Correlation between Variables',size=18)
plt.xticks(size=13)
plt.yticks(size=13)
```
